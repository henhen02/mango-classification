# Pemrosesan Gambar
Tahap ini melibatkan proses untuk pemampatan citra. Tujuan dari proses ini adalah untuk memperkecil ukuran citra demi mengurangi beban komputasi pada komputer dan juga untuk menyelaraskan setiap data menjadi ukuran yang sama.

```python
import cv2
import os
```

```python

def load_dir(data_dir, new_dir):
    path = []
    new_path = []
    for sub_dir in os.listdir(data_dir):
        path.append(os.path.join(data_dir, sub_dir))
        new_dir_path = os.path.join(new_dir, sub_dir)
        new_path.append(new_dir_path)
        if not os.path.exists(new_dir_path):
            os.makedirs(new_dir_path)
    for i in range(len(path)):
        for image in os.listdir(path[i]):
            img = cv2.imread(os.path.join(path[i], image))
            img = cv2.resize(img, (150, 150))
            cv2.imwrite(os.path.join(new_path[i], image), img)
            print('Image {} saved'.format(image))
    print('All images saved')


load_dir('mangga_original_images', 'mangga_150x150')
            
```
